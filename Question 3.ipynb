```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
months=np.array(["January","February","March","April"])
attandance=np.array([85,90,95,88])

plt.plot(months,attandance)

plt.title("Attandance Graph")
plt.xlabel("Months")
plt.ylabel("Attandance")
plt.show()
```

```python
# Attandance has increased rapidly then falls down steadily.
```

```python
# In April attandance has dropped to 88 from highest point 95 in March
```
