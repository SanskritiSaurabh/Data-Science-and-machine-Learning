```python
import matplotlib.pyplot as plt
months=['jan','Feb','Mar','Apr']
sales=[110,120,130,90]
plt.plot(months,sales)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
mon=['Jan','Feb','March','April']
att=[85,90,95,88]
plt.plot(mon,att)
plt.xlabel("Months")
plt.ylabel("Attandence")
```

```python
dire=['North','South','East','West']
rev=[200000,150000,180000,220000]
plt.bar(dire,rev)
plt.title("Revenue Chart by Region")
plt.xlabel("Direction")
plt.ylabel("Revenue")
plt.show()
```

```python
direc=['North','South','East','West']
book=[2000,1500,1800,2200]
plt.bar(direc,book,color=["Grey","Pink","yellow","Aqua"])
plt.title("Books sold Chart by Wing")
plt.xlabel("Direction")
plt.ylabel("Books")
plt.show()
```

```python

fig,ax=plt.subplots(figsize=(3,4))
months=['Jan','Feb','March','April','May','June']
sales=[4200,5500,4900,6700,7100,8300]
ax.plot(months,sales,color="Red",linewidth=2.5,marker='p',markersize=7)
ax.set_title("Monthly sale")
fig,ax1=plt.subplots(figsize=(3,4))
ax1.plot(months,sales,color="Red",linewidth=2.5,marker='p',markersize=7)
#fig.plot(months,sales,color="Blue",linewidth=2.5,marker='x',markersize=7)
```

```python
ad_spend=np.array([200,400,600,800,1000,120,1400,700])
sales=np.array([20,30,19,39,60,40,100,89])

z=np.polyfit(ad_spend,sales,1)
p=np.poly1d(z)
plt.plot(ad_spend,p(ad_spend),'r--',label='Trend')
```
