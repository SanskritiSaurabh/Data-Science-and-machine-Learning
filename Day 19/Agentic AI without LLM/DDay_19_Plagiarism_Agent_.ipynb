Scenario: The University’s AI Plagiarism Agent
Imagine a university has deployed an AI-powered plagiarism detection agent named PlagiarismLLM.
Its job is to monitor student essay submissions in real time and decide whether they are original or potentially plagiarized.

Step 1: Thinking (Reasoning)
Whenever an essay is submitted, PlagiarismLLM first analyzes the text:
- Example: “Essay contains large verbatim passages from online sources”
- PlagiarismLLM thinks: “Analyzing essay: multiple verbatim passages detected.”

Step 2: Planning (Choosing an Action)
Based on the thought, PlagiarismLLM decides which tool to use:
- If the essay has long verbatim matches or suspicious similarity scores, it plans to flag as potential plagiarism.
- Otherwise, it plans to mark as original work.

Step 3: Acting (Running Tools)
PlagiarismLLM hands over the decision to PlagiarismTools, which executes the action:
- flag_plagiarism → returns “Potential Plagiarism Detected”
- mark_original → returns “Original Work”

Step 4: Answering (Final Decision)
Once a clear classification is reached, PlagiarismLLM announces the decision:
- “Decision: Potential Plagiarism Detected”
- “Decision: Original Work”

Step 1 — Tools

```python
def flag_plagiarism():
    return "Potential Plagiarism Detected"

def mark_original():
    return "Original Work"
```

Step 2 — Thinking (Reasoning Engine)

```python
def analyze_essay(essay_text):
    suspicious_phrases = [
        "according to wikipedia",
        "as stated in research paper",
        "lorem ipsum",
        "copied content",
        "verbatim"
    ]

    thought = ""

    # Check for suspicious phrases
    matches = [phrase for phrase in suspicious_phrases if phrase in essay_text.lower()]

    if len(matches) >= 2:
        thought = "Essay contains multiple verbatim or suspicious phrases."
        score = "high"
    elif len(matches) == 1:
        thought = "Essay contains some suspicious phrasing."
        score = "medium"
    else:
        thought = "Essay appears original with no major matches."
        score = "low"

    print(f"\n[Thinking] {thought}")
    return score
```

Step 3 — Planning (Decision Logic)

```python
def agent_planner(score):
    plan = []

    if score == "high":
        plan.append("flag")
    elif score == "medium":
        plan.append("flag")
    else:
        plan.append("original")

    print(f"\n[Plan] {plan}")
    return plan
```

Step 4 — Acting (Execution Engine)

```python
def execute_plan(plan):
    result = None

    for step in plan:
        if step == "flag":
            print("\n[Action] Flagging plagiarism...")
            result = flag_plagiarism()

        elif step == "original":
            print("\n[Action] Marking as original...")
            result = mark_original()

    return result
```

Step 5 — Full Agent (End-to-End Flow)

```python
def plagiarism_agent(essay_text):
    print("📄 New Essay Submitted")

    # 1. Thinking
    score = analyze_essay(essay_text)

    # 2. Planning
    plan = agent_planner(score)

    # 3. Acting
    decision = execute_plan(plan)

    # 4. Final Answer
    final_output = f"\nDecision: {decision}"
    return final_output
```

Step 6- Test Cases

```python
# Test Case 1 (Plagiarized)
essay1 = """
This essay is copied content. According to Wikipedia, AI is transforming the world.
As stated in research paper, machine learning is powerful.
"""

print(plagiarism_agent(essay1))


# Test Case 2 (Original)
essay2 = """
Artificial intelligence is an evolving field that enables machines to learn from data.
It has applications in healthcare, education, and automation.
"""

print(plagiarism_agent(essay2))
```
