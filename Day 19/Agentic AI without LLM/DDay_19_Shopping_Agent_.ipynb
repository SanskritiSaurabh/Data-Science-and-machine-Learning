Scenario: Smart Shopping Assistant 🛒

“Imagine you are planning a party and you tell your assistant:

‘Find me the best chocolate cake under ₹500.’

Now instead of just replying with text, your assistant actually thinks, checks different bakery websites, compares prices, and then gives you the final decision.”

Step 1: Define Tools

```python
def search_cakes():
    return [
    {"name": "Choco Delight", "price": 450, "rating": 4.5, "website": "BakeHub"},
    {"name": "Dark Fantasy Cake", "price": 550, "rating": 4.8, "website": "SweetWorld"},
    {"name": "Chocolate Truffle", "price": 480, "rating": 4.7, "website": "CakeZone"},
    {"name": "Budget Choco Cake", "price": 300, "rating": 4.0, "website": "LocalBakes"},
    {"name": "Premium Belgian Chocolate Cake", "price": 700, "rating": 4.9, "website": "LuxuryCakes"},
    {"name": "Classic Chocolate Cake", "price": 400, "rating": 4.3, "website": "BakeHub"}

    ]

def filter_budget(cakes, max_price):
    return [cake for cake in cakes if cake["price"] <= max_price]

def choose_best_cake(cakes):
    # scoring logic (rating high + price low)
    best = max(cakes, key=lambda x: (x["rating"], -x["price"]))
    return best
```

Step 2 — Agent Brain

```python
def agent_brain(user_query):
    plan = []

    if "cake" in user_query.lower():
        plan.append("search_cakes")
        plan.append("filter_budget")
        plan.append("choose_best")

    return plan
```

Step 3 — Execution Engine

```python
def execute_plan(plan, budget):
    context = {}

    for step in plan:

        if step == "search_cakes":
            print("\n[Action] Searching Cakes...")
            context["cakes"] = search_cakes()
            print(context["cakes"])

        elif step == "filter_budget":
            print("\n[Action] Filtering by Budget...")
            context["filtered"] = filter_budget(context["cakes"], budget)
            print(context["filtered"])

        elif step == "choose_best":
            print("\n[Action] Choosing Best Cake...")
            best = choose_best_cake(context["filtered"])
            context["best_cake"] = best
            print(best)

    return context
```

Step 4 — Full Agent

```python
def shopping_agent(user_query, budget):
    print("User Query:", user_query)

    # Step 1: Plan
    plan = agent_brain(user_query)
    print("\n[Plan]", plan)

    # Step 2: Execute
    result = execute_plan(plan, budget)

    # Step 3: Final Output
    final_output = f"""
Best Cake Recommendation:
Name: {result['best_cake']['name']}
Price: ₹{result['best_cake']['price']}
Rating: {result['best_cake']['rating']}
Website: {result['best_cake']['website']}
Recommendation: Best value for money under ₹{budget}
"""
    return final_output
```

Step 5 — Run

```python
response = shopping_agent("Find me the best chocolate cake", 500)
print("\nFinal Answer:\n", response)
```
