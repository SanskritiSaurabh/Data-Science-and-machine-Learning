Scenario: Automated Business Report Assistant

🎯 Problem Statement

In companies:
- Managers spend hours compiling data from multiple sources (CRM, ERP, spreadsheets).
- Reports often require manual formatting, chart creation, and consistency checks.
- This slows down decision-making and introduces human error.

👉 Solution

A multi-agent AutoGen system where:

- 🤖 Assistant Agent drafts report templates and queries relevant data.
- ⚙️ UserProxy Agent executes data pulls and formatting automatically.
- 🔁 The system iterates until the report meets formatting, accuracy, and completeness standards.

🔄 Workflow

Business Request → AI drafts report structure → Data fetched & formatted → Charts auto-generated → Validation loop → Final polished report

```python
import os
import autogen
from google.colab import userdata


groq_api_key = userdata.get("newapi")
os.environ["GROQ_API_KEY"] = groq_api_key

llm_config = {
    "config_list": [
        {
            "model": "llama-3.3-70b-versatile",
            "api_key": groq_api_key,
            "base_url": "https://api.groq.com/openai/v1"
        }
    ],
    "temperature": 0
}


# Agent 1: Report agent

coder = autogen.AssistantAgent(
    name="Reporter",
    llm_config=llm_config,
    system_message="""You are an expert report writer.

drafts report templates and queries relevant data.

Rules:
- Check the report
- Add detailed structure
- Ensure the report is structered and precise
- End your response with DONE"""
)



# Agent 2: user proxy

reviewer = autogen.UserProxyAgent(
    name="Reviewer",
    human_input_mode="ALWAYS",
    max_consecutive_auto_reply=2,
    code_execution_config={
        "work_dir": "review",
        "use_docker": False
    },
    is_termination_msg=lambda x: "DONE" in x.get("content", ""),
    system_message="""You are a bussiness manager.

Responsibilities:
1. compiling data from multiple sources (CRM, ERP, spreadsheets)
2. format the report
3. check the correctness
4. ask to update the report and enhance it
5. Approve when correct"""
)


# Start Collaboration

reviewer.initiate_chat(
    coder,
    message="""Write a bussiness report:
- generate the structured report
- return the structured and detailed report
- generate query related data
"""
)
```

```python


import os
import autogen
from google.colab import userdata


os.environ["GROQ_API_KEY"] = userdata.get("newapi")
config_list = [
    {
        "model": "llama-3.3-70b-versatile",   # Groq model
        "api_key": os.environ["GROQ_API_KEY"],
        "base_url": "https://api.groq.com/openai/v1"
    }
]
llm_config = {"config_list": config_list, "temperature": 0}

#Assistant Agent
assistant = autogen.AssistantAgent(
    name="Data_Analyst",
    llm_config=llm_config,
    system_message="Write Python code for business analysis and explain results."
)

#Executor Agent
user_proxy = autogen.UserProxyAgent(
    name="Executor",
    human_input_mode="NEVER",   # No manual input, runs automatically
    max_consecutive_auto_reply=5,  # Limit auto-replies to 5 loops
    code_execution_config={"work_dir": "coding", "use_docker": False}
)

#Run Task (Agent-to-Agent Conversation)
user_proxy.initiate_chat(
    assistant,
    message="""
Analyze sales data: [12000, 15000, 10000, 18000, 13000]

Find:
- Mean
- Median
- Mode
- Highest & Lowest
- Business insights
"""
)
```

Scenario: AI-Powered Dev Workflow (Corporate Engineering Team)

 🎯 Problem Statement

 In companies:
 - Developers manually write and test code.
 - Reviewers check for bugs and compliance.
 - Testing and validation are slow and error-prone.

 👉 Solution

 A multi-agent system powered by Groq API where:
 - 👨‍💻 Coder Agent writes Python code for requested features.
 - 🔍 Reviewer Agent validates logic and ensures best practices.
 - 🧪 Tester Agent generates unit tests and executes them automatically.
 - 🔁 The loop continues until the code passes all tests and meets requirements.

 🔄 Workflow

 Feature Request → Product Manager breaks into tasks → Coder writes code → Tester validates with unit tests → Reviewer checks compliance → Iteration loop → Final approved codebase

 🧹 Conflict Resolution

 - Resolve conflicting dependencies (e.g., package version mismatches).
- Standardize coding style across agents (PEP8 compliance).
 - Ensure Groq API integration avoids latency or model mismatch issues.

```python
import os
import autogen
from google.colab import userdata

groq_api_key = userdata.get("newapi")
os.environ["GROQ_API_KEY"] = groq_api_key

llm_config = {
    "config_list": [
        {
            "model": "llama-3.3-70b-versatile",
            "api_key": groq_api_key,
            "base_url": "https://api.groq.com/openai/v1"
        }
    ],
    "temperature": 0
}


# Specialized agents
pm = autogen.AssistantAgent(
    name="ProductManager",
    llm_config=llm_config,
    system_message="You break down feature requests into tasks."
)

coder = autogen.AssistantAgent(
    name="Coder",
    llm_config=llm_config,
    system_message="You write Python code for given tasks."
)

tester = autogen.AssistantAgent(
    name="Tester",
    llm_config=llm_config,
    system_message="You write unit tests for code."
)

user = autogen.UserProxyAgent(
    name="User",
    human_input_mode="NEVER",
    code_execution_config={"work_dir": "project"}
)

# Group chat setup
groupchat = autogen.GroupChat(
    agents=[user, pm, coder, tester],
    messages=[],
    max_round=12,
    speaker_selection_method="auto"
)

# Manager orchestrates workflow
manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config
)

# Start the group task
user.initiate_chat(manager,
    message="Build a simple REST API for a todo app")
```

Scenario: AI-Powered HR Onboarding System (Corporate HR Team)

🎯 Problem Statement

In companies:
- HR teams manually prepare onboarding documents and checklists.
- Managers validate compliance with policies.
- Employees often face delays in receiving resources and training schedules.

👉 Solution

A multi-agent system where:
- 📝 HR Agent prepares onboarding documents, training schedules, and resource lists.
- ✅ Compliance Agent reviews documents for policy adherence and legal requirements.
- 📂 Resource Agent allocates accounts, devices, and access permissions.
- 🔁 The loop continues until all onboarding steps are validated and complete.

🔄 Workflow

Onboarding Request → HR Agent drafts documents → Compliance Agent validates → Resource Agent provisions accounts/devices → Iteration loop → Final approved onboarding package

🧹 Conflict Resolution

- Resolve conflicting policy requirements (e.g., different regional compliance rules).
- Standardize onboarding templates across departments.
- Ensure system avoids duplicate resource allocation (e.g., multiple accounts for same employee).

```python
import os
import autogen
from google.colab import userdata

# API Setup
groq_api_key = userdata.get("newapi")
os.environ["GROQ_API_KEY"] = groq_api_key

llm_config = {
    "config_list": [
        {
            "model": "llama-3.3-70b-versatile",
            "api_key": groq_api_key,
            "base_url": "https://api.groq.com/openai/v1"
        }
    ],
    "temperature": 0
}

# HR Agent
hr_agent = autogen.AssistantAgent(
    name="HR_Agent",
    llm_config=llm_config,
    system_message="""
You are an HR onboarding specialist.

Responsibilities:
- Prepare onboarding documents
- Create training schedules
- Generate employee resource checklist
- Follow standard onboarding templates

Output structured content clearly.
"""
)

# Compliance Agent
compliance_agent = autogen.AssistantAgent(
    name="Compliance_Agent",
    llm_config=llm_config,
    system_message="""
You are a compliance officer.

Responsibilities:
- Review onboarding documents
- Ensure legal and company policy adherence
- Check regional compliance requirements
- Highlight issues or approve

Respond with:
- APPROVED if correct
- Else provide corrections
"""
)

# Resource Agent
resource_agent = autogen.AssistantAgent(
    name="Resource_Agent",
    llm_config=llm_config,
    system_message="""
You are an IT/resource provisioning specialist.

Responsibilities:
- Allocate system accounts (email, tools)
- Assign devices (laptop, access cards)
- Grant permissions and access rights
- Avoid duplicate resource allocation

Confirm once provisioning is complete.
"""
)

# User / Orchestrator
user = autogen.UserProxyAgent(
    name="HR_Manager",
    human_input_mode="NEVER",
    code_execution_config={"work_dir": "hr_onboarding"},
)

# Group Chat Setup
groupchat = autogen.GroupChat(
    agents=[user, hr_agent, compliance_agent, resource_agent],
    messages=[],
    max_round=12,
    speaker_selection_method="auto"
)

# Manager
manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config
)

#Start Workflow
user.initiate_chat(
    manager,
    message="""
New employee onboarding request:

- Role: Software Engineer
- Location: India
- Department: IT

Tasks:
1. Prepare onboarding documents and training schedule
2. Validate compliance with company and regional policies
3. Allocate resources (email, system access, laptop)
4. Ensure no duplication and full completion

Continue discussion until onboarding is complete and approved.
"""
)
```
