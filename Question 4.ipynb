```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
salaries=np.random.normal(50000,10000,1000)
```

```python
print("Lowest Salary: ",round(salaries.min(),1))
print("Highest Salary: ",round(salaries.max(),1))
print("Average Salary: ",round(salaries.mean(),1))
```

```python
plt.hist(salaries,bins=20,color="red",edgecolor="white")
plt.axvline(salaries.mean(),color="yellow",linewidth=2,label='Average')
```
