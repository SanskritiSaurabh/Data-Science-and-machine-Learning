 Scenario: AI Content Automation for a Startup

 рџЏў Problem Statement

 A startup wants to:

 Stay updated with latest AI trends

 Automatically create LinkedIn posts

 Save time on manual research & writing

So we build a multi-agent AI system to automate this using CrewAI.

```python
!pip install crewai litellm langchain-community langchain-litellm
```

```python
import os
os.environ["CREWAI_TRACING_ENABLED"] = "true"
from crewai import Agent, Task, Crew
from google.colab import userdata
from langchain_litellm import ChatLiteLLM

# 1. Set Groq API Key
os.environ["GROQ_API_KEY"] = userdata.get("newapi")


if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]


# 2. Define LLM (using ChatLiteLLM for Groq)
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# 3. Define Agents
researcher = Agent(
    role="Researcher",
    goal="Find latest AI news",
    backstory="Expert in AI trends",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

writer = Agent(
    role="Writer",
    goal="Write LinkedIn post",
    backstory="Expert content creator",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)


# 4. Tasks
research_task = Task(
    description="Find 5 latest AI news trends",
    expected_output="List with summaries",
    agent=researcher
)

write_task = Task(
    description="Write LinkedIn post",
    expected_output="Final post",
    agent=writer,
    context=[research_task]
)

# 5. Crew
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
    verbose=True
)


# 6. Run
result = crew.kickoff()

print("\nFINAL OUTPUT:\n")
print(result)
```

Scenario: Smart Job Application Assistant

 рџЏў Problem Statement

 Students struggle to:

 Analyze job descriptions

 Match their skills

 Write tailored resumes & cover letters

 So we build a multi-agent AI system that automates job application preparation.

 рџ¤– System Overview
 рџ”№ Agent 1: Job Analyzer

 Reads job description

 Extracts required skills

 рџ”№ Agent 2: Resume Optimizer

 Matches candidate skills with job

 Suggests improvements

 рџ”№ Agent 3: Cover Letter Writer

 Generates a personalized cover letter

 рџ”„ Workflow
Job Description в†’ Skill Analysis в†’ Resume Optimization в†’ Cover Letter Generation

```python
!pip install crewai litellm
```

```python
import os
from crewai import Agent, Task, Crew
from google.colab import userdata


# 1. API Key Setup
os.environ["GROQ_API_KEY"] = userdata.get("newapi")


# 2. Define Agents
job_analyzer = Agent(
    role="Job Analyzer",
    goal="Extract key skills and requirements from job descriptions",
    backstory=(
        "You are an HR expert who understands job postings and identifies "
        "important skills, qualifications, and expectations."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

resume_optimizer = Agent(
    role="Resume Optimizer",
    goal="Align candidate resume with job requirements",
    backstory=(
        "You are a career coach who improves resumes by matching them "
        "with job descriptions and highlighting relevant skills."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

cover_letter_writer = Agent(
    role="Cover Letter Writer",
    goal="Write a personalized cover letter",
    backstory=(
        "You are a professional writer who crafts compelling cover letters "
        "tailored to specific job roles."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)


# 3. Input Data (Scenario)
job_description = """
We are looking for a Data Analyst with experience in Python, SQL, Power BI,
and data visualization. The candidate should be able to analyze datasets,
create dashboards, and communicate insights effectively.
"""

candidate_resume = """
I am a BTech student skilled in Python and basic SQL.
I have created Power BI dashboards and completed projects on data analysis.
"""


# 4. Tasks
analysis_task = Task(
    description=f"""
    Analyze the following job description and extract:
    - Required skills
    - Key responsibilities
    - Tools/technologies

    Job Description:
    {job_description}
    """,
    expected_output="Structured list of skills, responsibilities, and tools",
    agent=job_analyzer
)

resume_task = Task(
    description=f"""
    Compare the candidate resume with the job requirements and:
    - Identify skill gaps
    - Suggest improvements
    - Optimize the resume

    Resume:
    {candidate_resume}
    """,
    expected_output="Improved resume with suggestions",
    agent=resume_optimizer,
    context=[analysis_task]
)

cover_letter_task = Task(
    description="""
    Create a personalized cover letter using:
    - Job requirements
    - Candidate resume
    - Improvements suggested

    Make it professional and engaging.
    """,
    expected_output="Final cover letter",
    agent=cover_letter_writer,
    context=[analysis_task, resume_task]
)


# 5. Crew Setup
crew = Crew(
    agents=[job_analyzer, resume_optimizer, cover_letter_writer],
    tasks=[analysis_task, resume_task, cover_letter_task],
    verbose=True,
    tracing=True
)

# 6. Run
result = crew.kickoff()

print("\n FINAL OUTPUT:\n")
print(result)
```

 Scenario: Automated Business Intelligence (BI) Report Generator

 рџЋЇ Problem Statement

 In companies:

 Managers struggle to analyze large sales data

 Reports take hours to prepare

 Insights are often delayed

 Solution: Build a multi-agent AI system that:

 Analyzes business data

 Extracts insights

 Generates executive reports

 System Overview

 рџ”№ Agent 1: Data Analyst

 Analyzes sales/business data

 Finds trends, patterns, KPIs

 рџ”№ Agent 2: Business Strategist

 Interprets insights

 Suggests business decisions

 рџ”№ Agent 3: Report Generator

 Creates a professional executive report

```python
import os
from crewai import Agent, Task, Crew
from google.colab import userdata


# 1. API Key Setup
os.environ["GROQ_API_KEY"] = userdata.get("newapi")


# 2. Define Agents
data_analyst = Agent(
    role="Data Analyst",
    goal="Analyze business data and extract meaningful insights",
    backstory=(
        "You are a senior data analyst working in a corporate environment. "
        "You specialize in identifying trends, KPIs, and patterns in sales data."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

business_strategist = Agent(
    role="Business Strategist",
    goal="Provide strategic recommendations based on data insights",
    backstory=(
        "You are a business consultant who translates data insights into "
        "actionable strategies for company growth."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

report_generator = Agent(
    role="Executive Report Writer",
    goal="Generate a professional business report",
    backstory=(
        "You are an expert in corporate communication who prepares executive-level "
        "reports with clarity and professionalism."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)


# 3. Sample Business Data
sales_data = """
Region: North | Revenue: 12000 | Cost: 8000
Region: South | Revenue: 15000 | Cost: 9000
Region: East  | Revenue: 10000 | Cost: 7000
Region: West  | Revenue: 18000 | Cost: 11000
"""


# 4. Tasks
analysis_task = Task(
    description=f"""
    Analyze the following sales data and provide:
    - Total revenue and profit
    - Best performing region
    - Key trends and observations

    Data:
    {sales_data}
    """,
    expected_output="Detailed analysis with KPIs and trends",
    agent=data_analyst
)

strategy_task = Task(
    description="""
    Based on the analysis:
    - Suggest 3 business strategies
    - Identify growth opportunities
    - Highlight risks
    """,
    expected_output="Strategic recommendations",
    agent=business_strategist,
    context=[analysis_task]
)

report_task = Task(
    description="""
    Create a professional executive report including:
    - Summary
    - Key insights
    - Recommendations
    - Conclusion
    """,
    expected_output="Well-structured business report",
    agent=report_generator,
    context=[analysis_task, strategy_task]
)


# 5. Crew Setup
crew = Crew(
    agents=[data_analyst, business_strategist, report_generator],
    tasks=[analysis_task, strategy_task, report_task],
    verbose=True,
    tracing=True
)



# 6. Run
result = crew.kickoff()

print("\nFINAL BUSINESS REPORT:\n")
print(result)
```
