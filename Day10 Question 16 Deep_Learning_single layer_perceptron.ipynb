```python
# Scenario:
# We want to classify whether a restaurant order is “Large” (1) or “Small” (0) based on the number
#  of items ordered.
# - If items ≥ 3 → Large order
# - If items < 3 → Small order
# 1 is large 0 is small
```

```python
# X = number of orders , y = large/small
import numpy as np
X = np.array([[2], [1], [6], [8],[2],[5],[1]])
y = np.array([0, 0, 1, 1,0,1,0]) 
```

```python
# Step 3: Activation Function (Step function)
def activation(z):
    return 1 if z >= 0 else 0
```

```python
weights = np.random.rand(1) 
bias = np.random.rand(1)
learning_rate = 0.2
```

```python
for epoch in range(10):  # run for 10 iterations
    print(f"Epoch {epoch+1}")
    for i in range(len(X)):
        # Linear combination
        z = np.dot(X[i], weights) + bias
        y_pred = activation(z)

        # Error
        error = y[i] - y_pred

        # Update rule
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}")
```

```python
print("\nFinal Weights:", weights)
print("Final Bias:", bias)
```

```python
# Step 5: Test the perceptron
test_items = np.array([1, 7])
for h in test_items:
    z = np.dot(h, weights) + bias
    print(f"Number of items: {h}, Prediction: {activation(z)}")
    if z==0:
        print('Resturaunt is Small')
    else:
        print('Resturaunt is large')
```
