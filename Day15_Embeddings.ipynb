```python
#  Scenario: University Library Search
# Imagine you’re building a smart search system for a university library.
# - Problem: Students often type queries like “fast dog running” or “canine sprinting”, but
# the library catalog has books titled “The dog runs fast”. Traditional keyword search misses these connections.
# - Solution: Use embeddings to capture semantic meaning instead of exact words.


from sentence_transformers import SentenceTransformer
model_name = "all-MiniLM-L6-v2"
model = SentenceTransformer(model_name)

texts = [
    "The dog runs fast",
    "A quick canine sprints",
    "Python is a programming language"
]

embeddings = model.encode(texts)

print(f"Model: {model_name}")
print(f"Embedding shape: {embeddings.shape}")
print(f"First vector sample: {embeddings[0][:10]}")
```

Scenario: Hospital Patient Record Search
- Problem: Doctors and nurses often type queries like “heart attack” or “chest pain emergency”, but the hospital’s electronic health record (EHR) system stores diagnoses as “acute myocardial infarction”. Traditional keyword search fails because medical staff use everyday language while records use clinical terminology.
- Solution: Use embeddings to capture semantic meaning. Embeddings understand that “heart attack” and “acute myocardial infarction” are related concepts. This allows the search system to return the right patient records even when the query wording doesn’t match the stored terminology.

```python
model_name = "all-mpnet-base-v2"
model = SentenceTransformer(model_name)

texts = [
    "Heart Attack",
    "Chest pain emergency",
    "acute myocardial infarction"
]

embeddings = model.encode(texts)

print(f"Model: {model_name}")
print(f"Embedding shape: {embeddings.shape}")
print(f"First vector sample: {embeddings[0][:10]}")
```

Scenario: E‑Commerce Product Search
- Problem: Shoppers often type queries like “cheap running shoes” or “affordable sneakers for jogging”, but the catalog lists products as “Budget-friendly athletic footwear”. Traditional keyword search misses these connections because customers use everyday language while product listings use marketing terms.
- Solution: Use embeddings to capture semantic meaning. Embeddings understand that “cheap running shoes” and “budget-friendly athletic footwear” are related concepts. This ensures customers find the right products even when their wording doesn’t match the catalog descriptions.

```python
model_name = "BAAI/bge-m3"
model = SentenceTransformer(model_name)

texts = [
    "cheap running shoes",
    "affordable sneakers for jogging",
    "Budget-friendly athletic footwear"
]

embeddings = model.encode(texts)

print(f"Model: {model_name}")
print(f"Embedding shape: {embeddings.shape}")
print(f"First vector sample: {embeddings[0][:10]}")
```
