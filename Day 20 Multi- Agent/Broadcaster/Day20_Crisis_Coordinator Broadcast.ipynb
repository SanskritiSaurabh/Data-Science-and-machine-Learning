Agent 1: Crisis Coordinator (Broadcaster)
- Broadcasts: “Data breach detected in customer database. Immediate response required.”
- Sends this to all other agents simultaneously.

🛡️ Agent 2: IT Security Agent
- Receives broadcast.
- Responds: “Isolate affected servers, patch vulnerabilities, start forensic analysis.”

📞 Agent 3: Communications Agent
- Receives broadcast.
- Responds: “Draft internal memo, prepare press release, notify stakeholders.”

💰 Agent 4: Finance Agent
- Receives broadcast.
- Responds: “Estimate financial impact, allocate emergency funds, review insurance coverage.”

👩‍⚖️ Agent 5: Legal Agent
- Receives broadcast.
- Responds: “Assess regulatory obligations, prepare compliance reports, advise on liability.”

👩‍💼 Agent 6: HR Agent
- Receives broadcast.
- Responds: “Brief employees, provide guidance on handling customer queries, ensure morale support.”

🧑‍⚖️ Agent 7: Decision Agent (Coordinator)
- Collects all responses.
- Integrates into a final crisis response plan:
“Servers isolated, communications prepared, financial impact assessed, compliance secured, employees briefed.”

````python
from groq import Groq
from google.colab import userdata
import json

api_key = userdata.get("newapi")

if not api_key:
    raise ValueError("API key not found in Colab Secrets")

client = Groq(api_key=api_key)

MODEL_NAME = "llama-3.3-70b-versatile"
````

````python
# HELPER FUNCTIONS
def clean_response(text):
    text = text.strip()
    if text.startswith("```"):
        text = text.replace("```json", "").replace("```", "").strip()
    return text


def safe_parse_json(text):
    try:
        return json.loads(text)
    except:
        return {"error": "Invalid JSON", "raw_output": text}


def call_llm(prompt):
    try:
        response = client.chat.completions.create(
            model=MODEL_NAME,
            messages=[
                {"role": "system", "content": "Return ONLY valid JSON. No explanation."},
                {"role": "user", "content": prompt}
            ],
            temperature=0
        )

        content = response.choices[0].message.content

        if not content or content.strip() == "":
            return {"error": "Empty response"}

        content = clean_response(content)
        return safe_parse_json(content)

    except Exception as e:
        return {"error": str(e)}
````

````python
# AGENT 1: CRISIS COORDINATOR (BROADCASTER)
def crisis_coordinator():
    print("\n[Crisis Coordinator] Broadcasting crisis alert...")

    message = "Data breach detected in customer database. Immediate response required."
    return message
````

````python
# AGENT 2: Security Agent
def it_security_agent(message):
    print("\n[IT Security Agent]")

    prompt = f"""
    Crisis message: {message}

    Provide response actions.

    Output JSON:
    {{
      "actions": []
    }}
    """
    return call_llm(prompt)
````

````python
# Agent 3: Communication Agent
def communications_agent(message):
    print("\n[Communications Agent]")

    prompt = f"""
    Crisis message: {message}

    Provide communication strategy.

    Output JSON:
    {{
      "actions": []
    }}
    """
    return call_llm(prompt)
````

````python
# Agent 4: Finance Agent
def finance_agent(message):
    print("\n[Finance Agent]")

    prompt = f"""
    Crisis message: {message}

    Provide financial actions.

    Output JSON:
    {{
      "actions": []
    }}
    """
    return call_llm(prompt)
````

````python
# Agent 5: Legal Agent
def legal_agent(message):
    print("\n[Legal Agent]")

    prompt = f"""
    Crisis message: {message}

    Provide legal actions.

    Output JSON:
    {{
      "actions": []
    }}
    """
    return call_llm(prompt)
````

````python
# Agent 6: HR agent
def hr_agent(message):
    print("\n[HR Agent]")

    prompt = f"""
    Crisis message: {message}

    Provide HR actions.

    Output JSON:
    {{
      "actions": []
    }}
    """
    return call_llm(prompt)
````

````python
# AGENT 7: DECISION AGENT
def decision_agent(message, responses):
    print("\n[Decision Agent] Integrating responses...")

    prompt = f"""
    Crisis message:
    {message}

    Agent responses:
    {json.dumps(responses, indent=2)}

    Create a final crisis response plan.

    Output JSON:
    {{
      "final_plan": "",
      "priority_actions": []
    }}
    """
    return call_llm(prompt)
````

````python
# MAIN SYSTEM (BROADCAST FLOW)

def crisis_multi_agent_system():
    print("=== CRISIS MANAGEMENT SYSTEM ===")

    # Step 1: Broadcast
    message = crisis_coordinator()

    # Step 2: All agents receive same message
    responses = {
        "it_security": it_security_agent(message),
        "communications": communications_agent(message),
        "finance": finance_agent(message),
        "legal": legal_agent(message),
        "hr": hr_agent(message)
    }

    # Step 3: Final decision
    final_decision = decision_agent(message, responses)

    return {
        "broadcast": message,
        "responses": responses,
        "final_decision": final_decision
    }


# RUN
result = crisis_multi_agent_system()

print("\n===== FINAL OUTPUT =====")
print(json.dumps(result, indent=2))
````
